```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def get_epoch_values(df, epoch):
    return df[df[f"test_accuracy_step/epoch_{epoch}"].notnull()].dropna(axis=1)

times = []
accuracies = []

for i in range(11, 17):
    df = pd.read_csv(f"../logs/default/version_{i}/metrics.csv")
    df = df.rename(columns={'gpu_id: 0/memory.used (MB)': 'gpu'})


    epochs = sorted({int(column.split("_")[-1]) for column in df.columns if "/epoch_" in column})
    clean_df = pd.DataFrame(columns=["time", "epoch", "step", "gpu", "test_accuracy"])
    for epoch in epochs:
        epoch_values = get_epoch_values(df, epoch)
        epoch_values = epoch_values.rename(columns={
            f"test_accuracy_step/epoch_{epoch}": "test_accuracy",
            f"time_step/epoch_{epoch}": "time"
        })
        epoch_values["epoch"] = epoch
        clean_df = pd.concat([clean_df, epoch_values[["time", "epoch", "step", "gpu", "test_accuracy"]]])

    clean_df["time"] = clean_df["time"] - clean_df.time.min()
    
    mean_df = clean_df.groupby("epoch").mean()
    max_df = clean_df.groupby('epoch').max()
    min_df = clean_df.groupby('epoch').min()
    times.append(max_df["time"].tolist())
    accuracies.append(mean_df["test_accuracy"].tolist())
```

```python
import numpy as np

ts = np.array(times)
acc = np.array(accuracies)
plt.plot(ts.T, 1-acc.T);
```
